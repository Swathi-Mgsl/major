# To Find out accuracy levels of various machine learning techniques & Building of Model.
-------------------------------------------------------------------------------------------------------------------------------------------
[such as Support Vector Machines (SVM), K-Nearest Neighbor (KNN), Decision Trees (DT) , Logistic Regression (LR) and Random Forest (RF)].

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier


df = pd.read_csv('D:/MSC/INTERSHIP/AI/cardio_train (1).csv', sep=';')

print(df.isnull().sum())

df = df.dropna()


X = df.drop(['cardio'], axis=1)
y = df['cardio']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
lifestyle_factors = ['smoke', 'alco', 'active']
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
for i, factor in enumerate(lifestyle_factors):
    sns.countplot(data=df, x=factor, hue='cardio', ax=axes[i])
    axes[i].set_title(f'{factor.capitalize()} vs Cardiovascular Disease')
plt.tight_layout()
plt.show()

corr_matrix = df.corr()
plt.figure(figsize=(14, 7))
sns.heatmap(corr_matrix, annot=True, fmt='.2f')
plt.title('Correlation Matrix of Features')
plt.show()


fig, axes = plt.subplots(1, 2, figsize=(14, 7))
sns.histplot(data=df, x='age', hue='cardio', multiple='stack', ax=axes[0])
axes[0].set_title('Age Distribution')
sns.histplot(data=df, x='cholesterol', hue='cardio', multiple='stack', ax=axes[1])
axes[1].set_title('Cholesterol Level Distribution')
plt.tight_layout()
plt.show()


fig, axes = plt.subplots(1, 3, figsize=(18, 6))
sns.boxplot(data=df, x='cardio', y='age', ax=axes[0])
sns.boxplot(data=df, x='cardio', y='cholesterol', ax=axes[1])
sns.boxplot(data=df, x='cardio', y='ap_hi', ax=axes[2])
plt.tight_layout()
plt.show()
```

```python
models = {
    'SVM': SVC(),
    'KNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'{name} Accuracy: {accuracy:.2f}')
    print(classification_report(y_test, y_pred))

best_model = RandomForestClassifier()
best_model.fit(X_train, y_train)


import joblib
joblib.dump(best_model, 'heart_disease_model.pkl')
```
